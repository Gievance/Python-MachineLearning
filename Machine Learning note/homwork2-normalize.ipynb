```python
arr=[200,300,400,600,1000]
```

```python
#-
def generalise_max_min(arr):
    max_value=max(arr)
    min_value=min(arr)
    #print(max_value,min_value)
    return [(i-min_value)/(max_value-min_value) for i in arr]
arr_out=generalise_max_min(arr)
print(arr_out)
```

```python
#z-score
import math 
#
def calculate_mean(arr):
    res=0
    sum=0
    for i in arr:
        sum+=i
    res=sum/len(arr)
    return res
#
def caclculate_standard(arr):
    res=calculate_mean(arr)
    sta=0
    for i in arr:
        sta+=(i-res)**2
    sta/=len(arr)
    return math.sqrt(sta)
mean=calculate_mean(arr)
sta=caclculate_standard(arr)
arr=[(i-mean)/sta for i in arr]
arr
```
