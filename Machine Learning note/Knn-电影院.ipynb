```python
# 
import math

# 
movie_data = {"": [45, 2, 9, ""],
              "": [21, 17, 5, ""],
              "3": [54, 9, 11, ""],
              "3": [39, 0, 31, ""],
              "": [5, 2, 57, ""],
              "3": [3, 2, 65, ""],
              "": [2, 3, 55, ""],
              "": [6, 4, 21, ""],
              "": [7, 46, 4, ""],
              "": [9, 39, 8, ""],
              "": [9, 38, 2, ""],
              "": [8, 34, 17, ""]}
# ::[23,3,17,?]

test=[23,3,17]
# 1. 
KNN_dist=[]

for key,value in movie_data.items():# 
    dist=math.sqrt(((test[0]-value[0])**2+(test[1]-value[1])**2+(test[2]-value[2])**2))
    KNN_dist.append([key,round(dist,2)])

print(f"KNN_dist:{KNN_dist}") # 
print('\n')
# 2. 
KNN_dist.sort(key=lambda dis:dis[1])

print(f"KNN_dist:{KNN_dist}") # 

#3.k

k=5
KNN_k=KNN_dist[:5]

label={}
for v in KNN_k:# 
    movie=movie_data[v[0]] 
    if movie[3] not in label:
        label[movie[3]]=0
    label[movie[3]]+=1
    
vote=sorted(label.items(),key=lambda a:a[1],reverse=True)
print('\n')
print(vote,vote[0][0],sep='\n')
```
